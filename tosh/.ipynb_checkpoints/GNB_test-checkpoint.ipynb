```python
import pandas as pd
```

```python
df = pd.read_csv('cleaned_2011.csv')
```

```python
df
```

```python
df['loan_status'].unique()
```

```python
# 1 good / 0 bad

df=df.replace(to_replace="Fully Paid",value= 1)
df=df.replace(to_replace="Charged Off",value= 0)
df=df.replace(to_replace="Current", value = 1)
df=df.replace(to_replace="Default",value= 0)
df=df.replace(to_replace="Late (31-120 days)",value= 0)
df=df.replace(to_replace="In Grace Period",value= 0)
df=df.replace(to_replace="Late (16-30 days)",value= 0)
```

```python
df['loan_status'].unique()
```

```python
df
```

```python
# get our y values for loan status and remove it from the data frame
y = df.pop('loan_status').values.reshape(-1,1)
```

```python
X = df.loc[:,:].values
```

```python
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
import numpy as np
```

```python
# split X and y into test and training sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.7, random_state = 1)
```

```python
# transform and normalize data
min_max_scaler = preprocessing.MinMaxScaler()
```

```python
X_train = min_max_scaler.fit_transform(X_train)
y_train = min_max_scaler.fit_transform(y_train)
```

```python
X_test = min_max_scaler.transform(X_test)
y_test = min_max_scaler.transform(y_test)
```

```python
# set up for Principle Component Analysis

from sklearn.decomposition import PCA
```

```python
pca = PCA(0.95)
```

```python
pca.fit(X_train)
```

```python
pca.n_components_
```

```python
pca.explained_variance_ratio_
```

```python
X_train_pca = pca.transform(X_train)
X_test_pca = pca.transform(X_test)
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
gnb = GaussianNB()
gnb.fit(X_train_pca, y_train)
```

```python
y_pred = gnb.predict(X_test_pca)
```

```python
from sklearn import metrics
```

```python
print("Gaussian NB model accuracy(%):", metrics.accuracy_score(y_test, y_pred)*100)
```
